```python
from tkinter import *
from tkinter import scrolledtext, Menu, filedialog, messagebox
from tkinter.ttk import Frame, Notebook
import pyperclip
```

```python
def tab():
        tab1 = Frame(tabmain)
        tabmain.add(tab1,text="Untitled")
        tabmain.pack(expand=1, fill = "both")
        scrl = scrolledtext.ScrolledText(tab1,height=200, width=200)
        scrl.pack()
        
def openfile():
    filedialog.askopenfile(filetypes = (("Text Documents ('*.txt')","*.txt"),("All Documents","*.*")))

def saveas():
    filedialog.asksaveasfile(filetypes = (("Text Documents ('*.txt')","*.txt"),("All Documents","*.*")))

def exitwindow():
    window.destroy()
     
def newwindow():
    window = Tk()
    window.title("Notepad")
    window.geometry("700x600")
    tabmain = Notebook(window)
    tabmain.pack(expand=1, fill= "both")
    tabm = Frame(tabmain)
    tabmain.add(tabm, text = "      Untitled")
    tabmain.pack(expand = 1, fill = "both")
    scrl = scrolledtext.ScrolledText(tabm,height=200, width=200)
    scrl.pack()
    btn = Button(window,text="+",command=tab)
    btn.place(x=0,y =0, width = 20, height= 23) 
    window.mainloop()
    
def cut():
    text = scrl.get("sel.first", "sel.last")
    window.clipboard_clear()
    window.clipboard_append(text)
    scrl.delete("sel.first", "sel.last")
    
def copy():
    text = scrl.get("sel.first", "sel.last")
    window.clipboard_clear()
    window.clipboard_append(text)

def paste():
    text = window.clipboard_get()
    scrl.insert("insert", text)
    
def delete():
    scrl.delete("sel.first", "sel.last")
    
def selectall():
    scrl.tag_add("sel","1.0","end")

def info():
    messagebox.showinfo("About","It's MN_13's Notepad")
    
def contact():
    messagebox.showinfo("Contact", "mnkv1013@gmail.com")
    
```

```python
window = Tk()
window.title("Notepad")
window.geometry("700x600")

tabmain = Notebook(window)
tabmain.pack(expand=1, fill= "both")
tabm = Frame(tabmain)
tabmain.add(tabm, text = "      Untitled")
tabmain.pack(expand = 1, fill = "both")

scrl = scrolledtext.ScrolledText(tabm,height=200, width=200)
scrl.pack()

btn = Button(window,text="+",command=tab)
btn.place(x=0,y =0, width = 20, height= 23) 


men = Menu(window)
new1 = Menu(men, tearoff = 0)
new1.add_command(label = "New Tab", command = tab)
new1.add_separator()
new1.add_command(label = "Open", command = openfile)
new1.add_command(label = "Save", command = saveas)
new1.add_command(label = "Save as", command = saveas)
new1.add_separator()
new1.add_command(label = "New Window", command = newwindow)
new1.add_separator()
new1.add_command(label = "Exit", command = exitwindow)

men.add_cascade(label = "File", menu = new1)


new2 = Menu(men, tearoff = 0)
new2.add_command(label = "Undo                 Ctrl+Z")
new2.add_separator()
new2.add_command(label = "Cut                    Ctrl+X", command=cut)
new2.add_command(label = "Copy                 Ctrl+C", command=copy)
new2.add_command(label = "Paste                 Ctrl+V", command=paste)
new2.add_command(label = "Delete                Del", command=delete)
new2.add_separator()
new2.add_command(label = "Find                  Ctrl+F" )
new2.add_command(label = "Find next            F3")
new2.add_command(label = "Find next          Ctrl+F3")
new2.add_command(label = "Repalace           Ctrl+H")
new2.add_command(label = "Go to                 Ctrl+G")
new2.add_separator()
new2.add_command(label = "Select All           Ctrl+A", command=selectall)
new2.add_command(label = "Date/Time            F5")

men.add_cascade(label = "Edit", menu = new2)


new3 = Menu(men, tearoff = 0)
new4 = Menu(men, tearoff = 0)
new4.add_command(label = "Zoom In                        Ctrl+Plus")
new4.add_command(label = "Zoom Out                    Ctrl+Minus")
new4.add_separator()
new4.add_command(label = "Restore to default             Ctrl+O")

new3.add_cascade(label = "Zoom", menu = new4)

new3.add_separator()
new3.add_checkbutton(label = "Status bar")
new3.add_checkbutton(label = "Word Wrap")

men.add_cascade(label = "View", menu = new3)

new5 = Menu(men, tearoff = 0)
new5.add_command(label = "About this Notebook", command = info)
new5.add_separator()
new5.add_command(label = "Contact", command = contact)

men.add_cascade(label = "Help", menu = new5)




window.config(menu = men)
window.mainloop()
```
